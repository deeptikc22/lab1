Maximum number


```python
def find_max(numbers):
    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num
nums = [2, 9, 0, 15, 40]
print("Maximum number:", find_max(nums))
```

Area of circle

```python
def area(r):
    pi = 3.14159
    area = pi * r * r
    return area

print("Area of circle:", area(5))
```

Area of rectangle

```python
def area(l,b):
    area= l * b
    return area

print("Area of rectangle:",area(5,4))
```

To find whether the entered character is vowel or not:

```python
def vowel(ch):
    return ch.lower() in "aeiou"
print(vowel('a'))
print(vowel('e'))
print(vowel('b'))
```
